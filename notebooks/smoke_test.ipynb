```python
mensagem = "default"
execucao = "manual"
```

```python
import io
import os
from datetime import datetime

print("Smoke test Papermill OK ->", mensagem, "| execucao:", execucao)

# Prova opcional de conectividade com o MinIO (grava um arquivo na Landing).
try:
    from minio import Minio

    endpoint = os.environ.get("MINIO_ENDPOINT", "http://minio:9000")
    endpoint = endpoint.replace("http://", "").replace("https://", "")
    client = Minio(
        endpoint,
        access_key=os.environ["MINIO_ROOT_USER"],
        secret_key=os.environ["MINIO_ROOT_PASSWORD"],
        secure=False,
    )
    bucket = os.environ.get("DATALAKE_BUCKET", "datalake")
    payload = f"smoke {datetime.utcnow().isoformat()} | {mensagem}".encode()
    client.put_object(bucket, "landing/_smoke_test.txt", io.BytesIO(payload), length=len(payload))
    print("Arquivo gravado no MinIO:", bucket, "/ landing/_smoke_test.txt")
except Exception as exc:  # noqa: BLE001 - smoke test nao deve falhar por MinIO
    print("Conexao com MinIO opcional falhou (ok para o smoke test):", exc)
```
